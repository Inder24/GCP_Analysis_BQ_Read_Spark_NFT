```python
from pyspark.sql import SparkSession
from pyspark.sql import SQLContext
from pyspark.sql.functions import *
from pyspark.sql.types import *
spark2 =SparkSession.builder \
  .master('local[*]') \
  .appName('conversions') \
  .config('spark.jars.packages', 'com.google.cloud.spark:spark-bigquery-with-dependencies_2.12:0.23.2') \
  .getOrCreate() 
sqlContext = SQLContext(spark2)
```

```python
df = sqlContext.read.format("csv").option("header","true").option("delimiter", ",").load("gs://nft_price_dataset_abhijit/Data_API.csv")
df.show(10)
```

```python
df.count()
```

```python
df.printSchema()
```

```python
df = df.drop("Image_url_1","Image_url_2","Image_url_3","Image_url_4")
```

```python
df.filter(df.Smart_contract.isNull()).show(10)
df.filter(df.ID_token.isNull()).show(10)
df.filter(df.Seller_address.isNull()).show(10)
df.filter(df.Seller_username.isNull()).show(10)
df.filter(df.Buyer_address.isNull()).show(10)
df.filter(df.Collection.isNull()).show(10)
df.filter(df.Collection_cleaned.isNull()).show(10)
df.filter(df.Market.isNull()).show(10)
```

```python
df1 = df.dropna()
```

```python
df1.count()
```

```python
df.select("Price_Crypto").show(15)
df1.select("Price_Crypto").show(15)
```

```python
df1.write.format('bigquery').option("temporaryGcsBucket","temp_bucket_temp").option('table', 'historical_data_set.nft_detail_data').mode('overwrite').save()
```
